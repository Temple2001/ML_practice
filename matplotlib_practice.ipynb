```
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x_data = np.random.rand(100)
y_data = np.random.rand(100)

plt.title('scatter plot') # 그래프 제목
plt.grid() # 그래프 좌표선
plt.scatter(x_data, y_data, color='b', marker='o') # 그래프 분포와 형태 설정
plt.show() # 그래프 보이기
```

```
import matplotlib.pyplot as plt

%matplotlib inline

x_data = [x for x in range(-5,5)]
y_data = [y*y for y in range(-5,5)]

plt.title('line plot')
plt.grid()
plt.plot(x_data, y_data, color='b')
plt.show()
```

```
# 함수 여러개 함께 그리기
import matplotlib.pyplot as plt

%matplotlib inline

x_data = [x for x in range(-4,5)]
y_data1 = [x*x for x in x_data]
y_data2 = [(x-1)*(x-1)+3 for x in x_data]

plt.title('line overwrite')
plt.grid()
plt.xlabel('x asis') # x 이름
plt.ylabel('y asis') # y 이름

plt.xlim(-4,4) # x 범위 제한
plt.ylim(0,20) # y 범위 제한

plt.plot(x_data, y_data1, label='x*x', color='b', lw=1.1, ls='-') # lw = line width, ls = line style
plt.plot(x_data, y_data2, label='(x-1)*(x-1)+3', color='g', lw=2.2, ls='--')

plt.legend(loc='best') # 함수 정보 표기

plt.show()
```

**Q) matplotlib을 이용해 x = np.arange(0,15,0.1), y1 = np.sin(x), y2 = np.cos(x)의 그래프를 출력하기**

```
import matplotlib.pyplot as plt

%matplotlib inline

x_data = np.arange(0,15,0.1)
y_data1 = np.sin(x_data)
y_data2 = np.cos(x_data)

plt.title('sin(x) / cos(x)')
plt.grid()
plt.xlabel('time')
plt.ylabel('amplitude')

plt.xlim(0,15)
plt.ylim(-1,1)

plt.plot(x_data, y_data1, label='sin(x)', color='r', ls='--')
plt.plot(x_data, y_data2, label='cos(x)', color='g', ls='--')

plt.legend(loc='best')

plt.show()
```
